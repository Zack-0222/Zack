# Group

```python
# creat a new column
data_europe['gdp_high_1962'] = data_europe['gdpPercap_1962'] > data_europe['gdpPercap_1962'].mean()
```

```python
data_europe
```

```python
data_europe.groupby('gdp_high_1962')
```

```python
data_europe.groupby('gdp_high_1962').mean()
```
